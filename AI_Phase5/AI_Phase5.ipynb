```python
import pandas as pd
import re
```

```python
## Load the data
```

```python
data = pd.read_csv('spam.csv')
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text = re.sub(r'[^a-z ]', '', text)  # Remove non-alphabetic characters
    return text
```

```python
data['v1'] = data['v1'].apply(clean_text)
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
```

## Tokenization and stopword removal

```python

vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(data['v1'])
# TF-IDF transformation
tfidf_transformer = TfidfTransformer()
X = tfidf_transformer.fit_transform(X)
from sklearn.preprocessing import LabelEncoder
```

```python

label_encoder = LabelEncoder()
data['v2'] = label_encoder.fit_transform(data['v2'])
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, data['v2'], test_size=0.2, random_state=42)
from sklearn.naive_bayes import MultinomialNB
```

```python
classifier = MultinomialNB()
classifier.fit(X_train, y_train)
from sklearn.metrics import accuracy_score, classification_report
```

```python
y_pred = classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python

print("Accuracy:", accuracy)
print(report)
```
